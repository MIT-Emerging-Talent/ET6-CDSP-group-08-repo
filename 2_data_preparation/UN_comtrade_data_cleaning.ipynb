```python
# Rename the columns for clarity
new_df.rename(
    columns={
        "refYear": "Year",
        "reporterISO": "Country_Code",
        "reporterDesc": "Country",
        "cmdCode": "Product_Code",
        "cmdDesc": "Product_Description",
        "netWgt": "Net_Weight_kg",
        "primaryValue": "Value_USD",
    },
    inplace=True,
)

# Display the first 5 rows of the DataFrame with the new column names
new_df.head()
```

```python
# Display the column names of the DataFrame after renaming
new_df.columns
```

```python
# check missing values
new_df.isna().sum()
```

```python
# Treat zero weights as missing
new_df["Net_Weight_kg"].value_counts()
```

```python
#  Treat zero weights as missing
new_df["Net_Weight_kg"] = new_df["Net_Weight_kg"].replace(0, np.nan)
```

```python
# check missing values after treating zero weights
new_df.isna().sum()
```

```python
# Compute average weight-to-value ratio per product
weight_value_ratio = (
    new_df[new_df["Net_Weight_kg"].notna() & new_df["Value_USD"] > 0]
    .groupby("Product_Description")
    .apply(lambda x: (x["Net_Weight_kg"] / x["Value_USD"]).mean())
    .to_dict()
)


# Define a function to estimate missing weight
def estimate_weight(row):
    if np.isnan(row["Net_Weight_kg"]):
        ratio = weight_value_ratio.get(row["Product_Description"])
        if ratio:
            return row["Value_USD"] * ratio
    return row["Net_Weight_kg"]


# Apply function to fill missing (zero) weights
new_df["Net_Weight_kg"] = new_df.apply(estimate_weight, axis=1)

# Check for missing values after estimation
new_df.isna().sum()
```

```python
conflict_years = {
    "Central African Rep.": list(range(2012, 2024)),
    "Somalia": list(range(1991, 2024)),
    "Pakistan": [
        2001,
        2004,
        2007,
        2008,
        2009,
        2014,
        2023,
        2025,
    ],  # Major conflict years
    "Mozambique": [2017, 2018, 2019, 2020, 2021],
    "Myanmar": list(range(2016, 2024)),
    "Ethiopia": [2020, 2021, 2022],
    "Ukraine": [2014, 2022, 2023],
    "Yemen": list(range(2014, 2024)),
    "Libya": [2011, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    "Afghanistan": list(range(2001, 2022)),
    "Nigeria": [2009, 2010, 2011, 2014, 2015, 2018, 2020],
    "Syria": list(range(2011, 2024)),
    "Iraq": list(range(2003, 2017)),
    "Sudan (...2011)": [2003, 2004, 2005, 2013, 2019, 2020, 2023, 2024, 2025],
    "Sudan": [2003, 2004, 2005, 2013, 2019, 2020, 2023, 2024, 2025],
    "South Sudan": list(range(2013, 2020)),
    "State of Palestine": list(range(2000, 2025)),
    "Mali": list(range(2010, 2024)),
}
```

```python
# Define a function to check if a year is a conflict year for a given country
def is_conflict_year(row):
    country = row["Country"]
    year = row["Year"]
    if country in conflict_years and year in conflict_years[country]:
        return True
    return False


new_df["Conflict_Year"] = new_df.apply(is_conflict_year, axis=1)
```
